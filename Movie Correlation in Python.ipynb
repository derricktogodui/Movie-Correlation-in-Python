```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# import the data

df = pd.read_csv(r'C:\Users\Derek Winters\Desktop\PROJECTS TO-DO\CORRELATION IN PYTHON\archive\movies.csv')
```

```python
# taking a look at the data

df.head()
```

```python
# remove missing values

df_cleaned = df.dropna()
```

```python
# for loop to find percentage of missing values in each column

for col in df.columns:
    percentage_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percentage_missing))
```

```python
# data types for columns

df.dtypes
```

```python
# changing the data type of the budget and gross columns from float to integer

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# create new column from 'released' column using the year only

df['released_yr_correct'] = df['released'].str.extract(r'(\d{4})')
```

```python
# sort the data frame by the gross column

df.sort_values(by='gross', inplace = False, ascending=False)
```

```python
# find and drop duplicates

df['company'].drop_duplicates().sort_values(ascending=False)

df.drop_duplicates()
```

```python
# scatter plot of budget vs gross revenue correlation

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('budget for film')
plt.ylabel('gross revenue')

plt.show()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"black"})
```

```python
# let us determine the correlation

df.corr(method='pearson')
```

```python
# visualize correlation in heatmap

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation matrix for numeric features')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show()
```

```python
# convert strings data type in company column into numbers to find correlation

df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# visualize correlation in heatmap after converting data types

correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation matrix for numeric features')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

high_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(high_pairs)
```
